```python
from pyspark.sql import functions as F
```

# METRO

**TABLA fact_metro**

**Objetivo: Crear una tabla hecho con los viajes de metro**

1) Crear tabla fact_metro


```python
fact_metro = (
    spark.table("hive_metastore.tfm.datos_brutos_metro")
    .select(
        "day",
        "station_complex_id",
        "total_ridership"
    )
    .withColumn("day", F.to_date("day"))
)

display(fact_metro.limit(10))
```

```python
#Por último, antes de guardar renomear nombres de columnas para que sea más sencillo entender las tablas

fact_metro = (
    fact_metro
    .withColumnRenamed('station_complex_id','metro_station_id')
)
```

```python
display(fact_metro.limit(10))
```

3) Guardar tabla fact_metro

```python
fact_metro.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("tfm.fact_metro")
```
