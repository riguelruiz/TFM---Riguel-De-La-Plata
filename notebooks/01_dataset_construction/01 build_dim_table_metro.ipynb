**Objetivo: Crear una tabla con ids unicos de l√≠neas de metro**

1) Crear tabla dim_metro

```python
dim_metro = spark.read.option("inferSchema", "true")\
    .option("header", "true")\
    .csv("abfss://datos@masterrdp001sta.dfs.core.windows.net/dim_mta_metro_ridership.csv")

dim_metro = (dim_metro).distinct()

display(dim_metro)
```

```python
dim_metro = (
    dim_metro
    .groupBy('station_complex_id_1', 'station_complex', 'borough')
    .agg(
        F.first("latitude").alias("latitude"),
        F.first("longitude").alias("longitude")
    )
)
```

2) A√±adir datos de la estacion meteorologica mas cercana

```python
station_weather = (spark.table("hive_metastore.tfm.dim_weather"))
```

```python
display(station_weather)
```

```python
#Cogiendo valores de latitud y longitud de las 4 estaciones meteorol√≥gicas

#EWR
latitude_ewr = (
    station_weather
    .select(F.col('latitude'))
    .where(F.col('station') == 'USW00014734')
    .first()['latitude']
)

longitude_ewr = (
    station_weather
    .select(F.col('longitude'))
    .where(F.col('station') == 'USW00014734')
    .first()['longitude']
)

#CENTRAL PARK
latitude_cp = (
    station_weather
    .select(F.col('latitude'))
    .where(F.col('station') == 'USW00094728')
    .first()['latitude']
)

longitude_cp = (
    station_weather
    .select(F.col('longitude'))
    .where(F.col('station') == 'USW00094728')
    .first()['longitude']
)

#LGA
latitude_lga = (
    station_weather
    .select(F.col('latitude'))
    .where(F.col('station') == 'USW00014732')
    .first()['latitude']
)

longitude_lga = (
    station_weather
    .select(F.col('longitude'))
    .where(F.col('station') == 'USW00014732')
    .first()['longitude']
)

#jfk
latitude_jfk = (
    station_weather
    .select(F.col('latitude'))
    .where(F.col('station') == 'USW00094789')
    .first()['latitude']
)

longitude_jfk = (
    station_weather
    .select(F.col('longitude'))
    .where(F.col('station') == 'USW00094789')
    .first()['longitude']
)

print(latitude_ewr)
print(longitude_ewr)

print(latitude_cp)
print(longitude_cp)

print(latitude_lga)
print(longitude_lga)

print(latitude_jfk)
print(longitude_jfk)
```

```python
#Calculando distancia a las estaciones meteorologicas

dim_metro = (
    dim_metro
    .withColumn(
        "distancia_ewr",
        2 * 6371 * F.asin(
            F.sqrt(
                F.sin((F.radians(F.lit(latitude_ewr)) - F.radians(F.col("latitude"))) / 2) ** 2 +
                F.cos(F.radians(F.col("latitude"))) *
                F.cos(F.radians(F.lit(latitude_ewr))) *
                F.sin((F.radians(F.lit(longitude_ewr)) - F.radians(F.col("longitude"))) / 2) ** 2
            )
        )
    )
    .withColumn(
        "distancia_cp",
        2 * 6371 * F.asin(
            F.sqrt(
                F.sin((F.radians(F.lit(latitude_cp)) - F.radians(F.col("latitude"))) / 2) ** 2 +
                F.cos(F.radians(F.col("latitude"))) *
                F.cos(F.radians(F.lit(latitude_cp))) *
                F.sin((F.radians(F.lit(longitude_cp)) - F.radians(F.col("longitude"))) / 2) ** 2
            )
        )
    )
    .withColumn(
        "distancia_lga",
        2 * 6371 * F.asin(
            F.sqrt(
                F.sin((F.radians(F.lit(latitude_lga)) - F.radians(F.col("latitude"))) / 2) ** 2 +
                F.cos(F.radians(F.col("latitude"))) *
                F.cos(F.radians(F.lit(latitude_lga))) *
                F.sin((F.radians(F.lit(longitude_lga)) - F.radians(F.col("longitude"))) / 2) ** 2
            )
        )
    )
    .withColumn(
        "distancia_jfk",
        2 * 6371 * F.asin(
            F.sqrt(
                F.sin((F.radians(F.lit(latitude_jfk)) - F.radians(F.col("latitude"))) / 2) ** 2 +
                F.cos(F.radians(F.col("latitude"))) *
                F.cos(F.radians(F.lit(latitude_jfk))) *
                F.sin((F.radians(F.lit(longitude_jfk)) - F.radians(F.col("longitude"))) / 2) ** 2
            )
        )
    )
)
```

```python
display(dim_metro)
```

```python
#Calculando estacion meteorologica mas cercana a la estacion de metro

dim_metro = (
    dim_metro
        .withColumn(
            "estacion_mas_cercana",
            F.array_sort(
                F.array(

                    F.struct(
                        F.col("distancia_ewr").alias("dist"),
                        F.lit("distancia_ewr").alias("nombre")
                    ),

                    F.struct(
                        F.col("distancia_cp").alias("dist"),
                        F.lit("distancia_cp").alias("nombre")
                    ),

                    F.struct(
                        F.col("distancia_lga").alias("dist"),
                        F.lit("distancia_lga").alias("nombre")
                    ),

                    F.struct(
                        F.col("distancia_jfk").alias("dist"),
                        F.lit("distancia_jfk").alias("nombre")
                    )

                )
            )[0]["nombre"]   
        )
)
```

```python
display(dim_metro)
```

```python
#A√±adiendo datos de id y eliminando columnas que ya no usar√©

dim_metro = (
    dim_metro
    .withColumn(
        'weather_station_id',
        F.when(F.col('estacion_mas_cercana') == 'distancia_ewr', 'USW00014734')
        .when(F.col('estacion_mas_cercana') == 'distancia_cp', 'USW00094728')
        .when(F.col('estacion_mas_cercana') == 'distancia_lga', 'USW00014732')
        .when(F.col('estacion_mas_cercana') == 'distancia_jfk', 'USW00094789')
    )
    .drop(
        "distancia_ewr",
        "distancia_cp",
        "distancia_lga",
        "distancia_jfk",
        'estacion_mas_cercana'
    )
)
```

```python
display(dim_metro)
```

```python
#Por √∫ltimo, antes de guardar renomear nombres de columnas para que sea m√°s sencillo entender las tablas

dim_metro = (
    dim_metro
    .withColumnRenamed('station_complex_id_1','metro_station_id')
    .withColumnRenamed('station_complex','metro_station_name')
)
```
