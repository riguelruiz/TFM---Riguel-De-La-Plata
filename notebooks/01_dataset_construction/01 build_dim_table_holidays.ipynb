# HOLIDAY

**TABLA dim_holiday**

**Objetivo: Crear una tabla con listado de días festivos**

```python
dim_holiday = spark.read.option("inferSchema", "true")\
    .option("header", "true")\
    .csv("abfss://datos@masterrdp001sta.dfs.core.windows.net/nyc_holidays_2020_2024.csv")
```

```python
from pyspark.sql.functions import col, date_sub
```

```python
display(dim_holiday)
```

Añadiendo información del día anterior al festivo

```python
dim_holiday = (
    dim_holiday
    .withColumn('date_d-1', date_sub(col('date'), 1))
)
```

```python
display(dim_holiday)
```

```python
dim_holiday.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("tfm.dim_holiday")
```
