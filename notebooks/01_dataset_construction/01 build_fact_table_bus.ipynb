```python
from pyspark.sql import functions as F
```

# BUS

**TABLA fact_bus**

**Objetivo: Crear una tabla hecho con los viajes de autobus**

1) Crear tabla fact_bus

```python
fact_bus = (
    spark.table("hive_metastore.tfm.datos_brutos_bus")
    .select(
        "day",
        "bus_route",
        "total_ridership"
    )
    .withColumn("day", F.to_date("day"))
)

display(fact_bus.limit(10))
```

2) Guardar tabla fact_bus

```python
fact_bus.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("tfm.fact_bus")
```
