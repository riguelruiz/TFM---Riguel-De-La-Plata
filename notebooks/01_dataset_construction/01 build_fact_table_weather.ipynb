# WEATHER

**Objetivo: Crear una tabla de hechos de clima.**

STATION	Código de la estación meteorológica.

NAME	Nombre de la estación.

DATE	Fecha de observación.

AWND	Velocidad promedio diaria del viento (mph).

PRCP	Precipitación total del día (pulgadas).

SNOW	Nieve caída durante el día (mm).

SNWD	Profundidad de nieve en el suelo (mm).

TMAX	Temperatura máxima diaria (fahrenheit).

TMIN	Temperatura mínima diaria (fahrenheit).

WT01	Niebla o neblina.

WT02	Tormenta eléctrica.

WT03	Lluvia o llovizna.

WT04	Nieve o aguanieve.

WT05	Granizo.

WT06	Nieve intensa o hielo.

WT08	Vientos fuertes o tormenta de viento.

1) Crear tabla fact_weather

```python
#Crear tabla
fact_weather_prev = (
    spark.table("hive_metastore.tfm.datos_brutos_weather")
    .select(
        "STATION",
        "DATE",
        "AWND",
        "PRCP",
        "SNOW",
        "SNWD",
        "TMAX",
        "TMIN",
        "WT01",
        "WT02",
        "WT03",
        "WT04",
        "WT05",
        "WT06",
        "WT08"
    )
)

display(fact_weather_prev.limit(10))
```

```python
#Poner todas las columnas en minisculo

for c in fact_weather_prev.columns:
    column = c.lower()
    fact_weather_prev = fact_weather_prev.withColumnRenamed(c, column)

display(fact_weather_prev.limit(100))
```

```python
fact_weather_prev1 = (
    fact_weather_prev
    .filter(F.col('station') == 'USW00094728')
)

fact_weather_prev2 = (
    fact_weather_prev
    .filter(F.col('station') != 'USW00094728')
)
```

```python
#Convertir temperatura de fahrenheit a celcius
fact_weather_prev2 = (
    fact_weather_prev2
    .withColumn('tmax', (F.col('tmax') - 32) * (5/9))
    .withColumn('tmin', (F.col('tmin') - 32) * (5/9))
)
```

```python
#Convertir pulgadas en mm en las columna prcp
fact_weather_prev2 = (
    fact_weather_prev2
    .withColumn("prcp", F.col("prcp") * 25.4)
)
```

```python
#Convertir mph en km/h
fact_weather_prev2 = (
    fact_weather_prev2
    .withColumn("awnd", F.col("awnd") * 1.609344)
)
```

```python
fact_weather = fact_weather_prev1.union(fact_weather_prev2)
```

```python
display(fact_weather.limit(10))
```

```python
#Por último, antes de guardar renomear nombres de columnas para que sea más sencillo entender las tablas

fact_weather = (
   fact_weather
    .withColumnRenamed('station','weather_station_id')
)
```
