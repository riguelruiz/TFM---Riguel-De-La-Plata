# BUS

**TABLA dim_bus**

**Objetivo: Crear una tabla con ids unicos de l√≠neas de autobuses**

1) Crear tabla dim_bus

```python
dim_bus = (
    spark.table("hive_metastore.tfm.datos_brutos_bus")
    .select("bus_route")
).distinct()

display(dim_bus)
```

2) Guardar tabla dim_bus en el catalogo

```python
dim_bus.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("tfm.dim_bus")
```
