```python
from pyspark.sql import functions as F
```

```python
dim_holiday = (spark.table("hive_metastore.tfm.dim_holiday"))

dim_holiday.cache()
```

1. Crear tabla calendario

```python
dim_calendar = (
    spark
    .range(1)
    .select(
        F.explode(
            F.sequence(
                F.to_date(F.lit('2020-01-01')),
                F.to_date(F.lit('2024-12-31')),
                F.expr('interval 1 day')
            )
        ).alias('date')
    )
)
```

```python
display(dim_calendar.limit(10))
```

2. JOIN con la tabla dim_holiday

```python
calendar = dim_calendar.alias('calendar')
hol = dim_holiday.alias('hol')
hol_d1 = dim_holiday.alias('hol_d1') 

dim_calendar = (
    calendar
    .join(
        hol,
        calendar['date'] == hol['date'],
        how = 'left'
    )
    .join(
        hol_d1,
        calendar['date'] == hol_d1['date_d-1'],
        how = 'left'
    )
    .select(
        F.col('calendar.date').alias('date'),
        F.col('hol.date').alias('holiday'),
        F.col('hol_d1.date_d-1').alias('holiday_d_1')
    )
)
```

```python
display(dim_calendar)
```

3. Categorizando das festivos

```python
dim_calendar = (
    dim_calendar
    .withColumn(
        'is_holiday',
        F.when(F.col('holiday').isNotNull(), 1).otherwise(0)
    )
    .withColumn(
        'is_pre_holiday',
        F.when(F.col('holiday_d_1').isNotNull(), 1).otherwise(0)
    ).drop('holiday', 'holiday_d_1')
)
```

```python
display(dim_calendar)
```

4. Fines de semana, das de la semana y mes

```python
#Fines de semana

dim_calendar = (
    dim_calendar
    .withColumn(
        'is_weekend',
        F.when((F.dayofweek('date') == 1) | (F.dayofweek('date') == 7), 1).otherwise(0)
    )
)
```

```python
display(dim_calendar)
```

```python
#Das de la semana

dim_calendar = (
    dim_calendar
    .withColumn(
        'day_of_week',
        F.dayofweek('date')
    )
)
```

```python
display(dim_calendar.limit(10))
```

```python
#Meses

dim_calendar = (
    dim_calendar
    .withColumn(
        'month',
        F.month('date')
    )
)
```

```python
display(dim_calendar.limit(10))
```

5. Guardar tabla

```python
dim_calendar.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("tfm.dim_calendar")
```
