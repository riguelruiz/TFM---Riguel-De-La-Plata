```python
from pathlib import Path
import pandas as pd

def apilar_csvs(pasta_csv: str, estandar: str = "*.csv") -> pd.DataFrame:
    
    archivos = sorted(Path(pasta_csv).glob(estandar))
    if not archivos:
        raise FileNotFoundError(f"Ninguno archivo encontrado en {pasta_csv} con estandar {estandar}")

    
    dfs = (
        pd.read_csv(
            arq,
            dtype_backend="pyarrow",  
            low_memory=False
        )
        for arq in archivos
    )
    df = pd.concat(dfs, ignore_index=True)
    return df
```

```python
pasta = r"C:\Riguel\Master\Clases\TFM\py\Metro"

df = apilar_csvs(pasta)
```
