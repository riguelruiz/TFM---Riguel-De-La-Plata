```python
from pathlib import Path
import pandas as pd

def empilhar_parquets(pasta_parquet: str, padrao: str = "*.parquet") -> pd.DataFrame:
    
    arquivos = sorted(Path(pasta_parquet).glob(padrao))
    if not arquivos:
        raise FileNotFoundError(
            f"Ning√∫n archivo encontrado {pasta_parquet}"
        )

    dfs = (
        pd.read_parquet(arq)
        for arq in arquivos
    )

    df = pd.concat(dfs, ignore_index=True)
    return df
```
